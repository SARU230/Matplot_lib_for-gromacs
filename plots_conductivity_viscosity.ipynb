Conductivity plot


```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, AutoMinorLocator

# ---------------- GLOBAL STYLE ----------------
plt.rcParams['font.family'] = 'sans'
plt.rcParams['axes.linewidth'] = 2.5

# ---------------- DATA ----------------
x_labels = [
    '80:20',
    '50:50',
    '20:80'
]
x = np.arange(len(x_labels))

data = {
    '10:100':  [7.02, 7.25, 7.93],
    '10:200':  [15.73, 16.7, 22.3],
    '100:500': [24.63, 20.77, 18.69],
    '200:500': [17.59, 24.68, 24.49],
    '250:1250':[24.66, 14.92, 16.94],
}

markers = ['s', 'o', '^', 'v', 'D']
colors = ['#4D4D4D', '#E41A1C', '#377EB8', '#4DAF4A', '#984EA3']

# ---------------- PLOT ----------------
fig, ax = plt.subplots(figsize=(8, 8))

for (label, y), m, c in zip(data.items(), markers, colors):
    ax.plot(
        x, y,
        marker=m,
        linewidth=3,
        markersize=10,
        color=c,
        label=label
    )

# ================= AXIS LIMIT OPTIONS =================
ax.set_xlim(-0.5, 2.5)      # change X limits here
ax.set_ylim(5, 50)         # change Y limits here

# ================= LABELS =================
ax.set_ylabel('Calculated Conductivity (mS\cm)', fontsize=20, labelpad=20)   # labelpad controls distance
ax.set_xlabel('(TEMPO:TEMPO$^+$)', fontsize=20, labelpad=20)

ax.set_xticks(x)
ax.set_xticklabels(x_labels, fontsize=16)

# ================= TICK LABEL DISTANCE =================
ax.tick_params(axis='x', pad=10)   # distance of x tick labels from axis
ax.tick_params(axis='y', pad=10)   # distance of y tick labels from axis

# ================= SPINES =================
for spine in ax.spines.values():
    spine.set_linewidth(2.5)

# ================= TICK CONTROL =================
ax.tick_params(axis='both', which='major',
               direction='in', length=10, width=2,
               labelsize=16, top=True, right=True)

ax.tick_params(axis='both', which='minor',
               direction='in', length=5, width=1.5,
               top=True, right=True)

ax.yaxis.set_major_locator(MultipleLocator(5))
ax.yaxis.set_minor_locator(MultipleLocator(1))
#ax.xaxis.set_minor_locator(AutoMinorLocator())

# ================= LEGEND =================
ax.legend(loc='upper right',
          bbox_to_anchor=(0.99, 0.98),
          fontsize=16,
          frameon=True,
          edgecolor='black')

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.ticker import MultipleLocator, AutoMinorLocator

# ---------------- GLOBAL STYLE ----------------
plt.rcParams['font.family'] = 'sans'
plt.rcParams['axes.linewidth'] = 2.5

# ---------------- DATA ----------------
x_labels = [
    '80:20',
    '50:50',
    '20:80'
]
x = np.arange(len(x_labels))

data = {
    '10:100':  [0.351, 0.350, 0.350],
    '10:200':  [0.400, 0.404, 0.415],
    '100:500': [0.438, 0.450, 0.453],
    '200:500': [0.460, 0.466, 0.476],
    '250:1250':[0.493, 0.504, 0.516],
}

markers = ['s', 'o', '^', 'v', 'D']
colors = ['#4D4D4D', '#E41A1C', '#377EB8', '#4DAF4A', '#984EA3']

# ---------------- PLOT ----------------
fig, ax = plt.subplots(figsize=(8, 8))

for (label, y), m, c in zip(data.items(), markers, colors):
    ax.plot(
        x, y,
        marker=m,
        linewidth=3,
        markersize=10,
        color=c,
        label=label
    )

# ================= AXIS LIMIT OPTIONS =================
ax.set_xlim(-0.5, 2.5)      # change X limits here
ax.set_ylim(0.2, 2.4)         # change Y limits here

# ================= LABELS =================
ax.set_ylabel('Calculated Viscosity (cP)', fontsize=20, labelpad=20)   # labelpad controls distance
ax.set_xlabel('(TEMPO:TEMPO$^+$)', fontsize=20, labelpad=20)

ax.set_xticks(x)
ax.set_xticklabels(x_labels, fontsize=16)

# ================= TICK LABEL DISTANCE =================
ax.tick_params(axis='x', pad=10)   # distance of x tick labels from axis
ax.tick_params(axis='y', pad=10)   # distance of y tick labels from axis

# ================= SPINES =================
for spine in ax.spines.values():
    spine.set_linewidth(2.5)

# ================= TICK CONTROL =================
ax.tick_params(axis='both', which='major',
               direction='in', length=10, width=2,
               labelsize=16, top=True, right=True)

ax.tick_params(axis='both', which='minor',
               direction='in', length=5, width=1.5,
               top=True, right=True)

ax.yaxis.set_major_locator(MultipleLocator(0.2))
ax.yaxis.set_minor_locator(MultipleLocator(1))
#ax.xaxis.set_minor_locator(AutoMinorLocator())

# ================= LEGEND =================
ax.legend(loc='upper right',
          bbox_to_anchor=(0.99, 0.98),
          fontsize=16,
          frameon=True,
          edgecolor='black')

plt.tight_layout()
plt.show()
```
